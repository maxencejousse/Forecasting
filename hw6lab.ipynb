```python
import pandas as pd
df = pd.read_csv("surnames_1990.csv")
```

`surnames_1990.csv` contains the frequency of the 5000 most common surnames in the US according to the 1990 Census. It is made of the first 5000 rows in [File B](https://www2.census.gov/topics/genealogy/2010surnames/names.zip) from [this website](https://www.census.gov/topics/population/genealogy/data/2010_surnames.html). 

Each row corresponds to a surname: the column `frequency` contains the percentage of responders with this surname, while the column `cumul_freq` contains the percentage of responders with this surname or a more common surname.

Q1. Plot the frequencies of the 200 most common surnames: for each surname, its rank should be on the x-axis, and its frequency should be on the y-axis.

```python
import matplotlib.pyplot as plt
import numpy as np 

plt.xlabel("Surname Rank")
plt.ylabel("Frequency")
plt.title("Frequency for the 200 most common surnames")
plt.plot(df.iloc[0:200, 1])
```

Q2. Plot the cumulative surname frequencies as a function of the surname frequencies on a log-log scale.

```python
df_log = df.copy()
df_log["frequency"] = np.log(df_log["frequency"])
df_log["cumul_freq"] = np.log(df_log["cumul_freq"])
plt.xlabel("Frequency")
plt.ylabel("Cumulative Frequency")
plt.plot(df_log["frequency"], df_log["cumul_freq"])
```

Q3. Based on the plots, what distribution do you think would be a good fit for the data? Explain your reasoning. 

A power distribution is a good candidate to model these data. Power law distributions are often observed in natural and social phenomena where a small number of events are extremely frequent, while the majority is rare.

Q4. Do you have a hypothesis as to why the data are distributed the way they are? Explain your reasoning in around 2-5 sentences.

This power-law distribution of surname frequencies can be accounted for by historical and social processes. With the passage of time, certain surnames became more frequent due to such factors as migration patterns, cultural rise, and events in the past. This would have resulted in a few extremely common surnames and many uncommon surnames. Additionally, the inheritance process of surnames from generation to generation is also causing this distribution to persist. Power law distribution is a result of the cumulative effect of these processes, and it results in some surnames being very frequent and many of them being rare.
